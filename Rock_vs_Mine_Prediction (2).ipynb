Importing the required dependicies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection

```python
#Loading the SONAR dataset into a pandas dataframe
sonar_data = pd.read_csv('/content/drive/MyDrive/Copy of sonar data (1).csv', header=None) #we have no header files here (no column names)
```

```python
#Printing the first five rows of the dataframe
sonar_data.head()
```

```python
#Number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
#Separating features and targets
x = sonar_data.drop(columns=60, axis=1)
y = sonar_data[60]
```

```python
print(x)
```

```python
print(y)
```

Splitting the dataset into training and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Training the model using Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

Model Evaluation

```python
#checking accuracy on the training data
x_train_prediction = model.predict(x_train)
train_accuracy = accuracy_score(x_train_prediction, y_train)

print('Training Data Accuracy: ', train_accuracy)
```

```python
#checking accuracy on the test data
x_test_prediction = model.predict(x_test)
test_accuracy = accuracy_score(x_test_prediction, y_test)

print('Test Data Accuracy: ', test_accuracy)
```

```python
#Visualizing Accuracy Plot

train_accuracy = model.score(x_train, y_train)
test_accuracy = model.score(x_test, y_test)

accuracies = [train_accuracy, test_accuracy]
labels = ['Training Accuracy', 'Test Accuracy']

plt.figure()
plt.bar(labels, accuracies)
plt.ylim(0,1)
plt.ylabel('Accuracy')
plt.title('Model Accuracy')
plt.show()
```
